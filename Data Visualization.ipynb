```python

import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns

# Load the CSV file
df = pd.read_csv("C:\\Users\\DELL\\OneDrive\\Desktop\\india_job_market_dataset.csv")
df

```

```python
sns.countplot(x="Job Title",data=df,hue="Job Title",palette='viridis')
plt.xticks(rotation=90)
plt.xlabel("Job Title",fontsize=12)
plt.ylabel("count",fontsize=14)
plt.title("count of roles by job titles",fontsize=13,color="green")
plt.show()
```

```python
gb=df.groupby("Job Type").agg(job_count=("Job ID", "count"))
gb
```

```python


# Group by "job type" and count the number of "job ID"
gb = df.groupby("Job Type").agg({"Job ID":"count"})

# Plot a pie chart
plt.pie(gb['Job ID'],labels=gb.index,autopct="%1.2f%%")
plt.title('Distribution of Jobs by Type')
plt.show()
```

```python
# Countplot with Job Title on x-axis and Salary Range as hue
sns.countplot(x="Job Title", data=df, hue="Salary Range", palette="spring")
plt.xticks(rotation=90)
plt.legend(bbox_to_anchor=(1,1))
plt.show()
```

```python
# Countplot with Job Title on x-axis and Salary Range as hue
sns.countplot(x="Job Title", data=df, hue="Salary Range", palette="GnBu")
plt.xticks(rotation=90)
plt.legend(bbox_to_anchor=(1,1))
plt.show()

```
